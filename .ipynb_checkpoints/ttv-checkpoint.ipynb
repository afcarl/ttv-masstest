```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def calc_D(sim, planet=1):
    D =(sim.particles[planet].y - sim.particles[0].y) * sim.particles[planet].vy 
    + (sim.particles[planet].z - sim.particles[0].z) * sim.particles[planet].vz
    return D
```

```python
def bisect(sim,t1,t2):
    tm = (t1+t2)/2.
    if t2-t1<(1.0e-8):
        return tm
    sim.integrate(tm)
    D = calc_D(sim, 1)
    if sim.particles[0].y>0.:
        t2 = tm
        return bisect(sim,t1,t2)
    else:
        t1 = tm
        return bisect(sim,t1,t2)
```

```python
sim = rebound.Simulation()
sim.add(m=1.)
planet1_params = {"a":0.5, "m":0.0002}
planet1_params = {"a":0.75, "m":0.0003}
sim.add(primary=sim.particles[0], **planet1_params)
#sim.move_to_com()
sim.integrate(1.)
```

```python
d_array = np.zeros(100)
for i in range(100):
    sim.step()
    d_array[i] = calc_D(sim, 1)
    
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.plot(np.arange(100), d_array)
plt.show()
print sim.t
```

```python
transits = 50
transits_var = 0
D_previous = -1.
time_previous = sim.t-0.01
transit_times = np.zeros(transits)
while(transits_var < transits):
    sim.step()
    D = calc_D(sim, 1)
    if(D>0 and D_previous<0):
        transit_times[transits_var] = bisect(sim, time_previous, sim.t)
        print sim.particles[1].xyz
        sim.integrate(sim.t + 0.5)
        transits_var += 1
    time_previous = sim.t
    D_previous = D
    
print transit_times
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(np.arange(transits),transit_times)
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.scatter(np.arange(transits), transit_times, marker="o")
lin = np.arange(transits)*slope + intercept
print lin
ax.plot(np.arange(transits), lin)
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
diff = (np.arange(transits)*slope + intercept) - transit_times
ax.plot(np.arange(transits), diff, marker="o")
plt.show()
```

```python
dt = np.zeros(len(transit_times)-1)
for i in range(len(dt)):
    dt[i] = transit_times[i] - transit_times[i+1]
print dt
```

```python
fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.plot(np.arange(transits-1), dt, marker="o")
plt.show()
```
